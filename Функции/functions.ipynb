Вы работаете секретарем и вам постоянно приходят различные документы. Вы должны быть очень внимательны, чтобы не потерять ни один документ. Каталог документов хранится в следующем виде:

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```

Перечень полок, на которых находятся документы хранится в следующем виде:

```python
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

Необходимо реализовать пользовательские команды (внимание! это не название функций, которые должны быть выразительными, а команды, которые вводит пользователь, чтобы получить необходимый результат):
  
- p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит;  
- s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится  
*Правильно обработайте ситуации, когда пользователь будет вводить несуществующий документ*.
- l – list – команда, которая выведет список всех документов в формате   
**passport "2207 876234" "Василий Гупкин"**  
- as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень. *Предусмотрите случай, когда пользователь добавляет полку, которая уже существует*.
- d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок. *Предусмотрите сценарий, когда пользователь вводит несуществующий документ*;  
- m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую. *Корректно обработайте кейсы, когда пользователь пытается переместить несуществующий документ или переместить документ на несуществующую полку*;  
- a – add – команда, которая добавит новый документ в каталог и в перечень полок, спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться. *Корректно обработайте ситуацию, когда пользователь будет пытаться добавить документ на несуществующую полку*.

```python

def check_num_doc():
    """Возвращает имя человека кому принадлежит введенный документ"""
    doc_num = input('Введите номер документа')
    i = 0
    for document in documents:
        if document['number'] == doc_num:
            print('Документ с введеным номером принадлежит человеку по имени',document['name'])
            i = 1
            break
    if i == 0:
        print('Документа с таким номером не найдено!')
            
            
```

```python
def check_shelf():
    """Возвращает номер полки на которой введенный документ"""
    doc_num = input('Введите номер документа')
    i = 0
    for shelf, document in directories.items():
        if doc_num in document:
            print('Документ с введеным номером лежит на полке:',shelf)
            i = 1
            break
    if i == 0:
        print('Документа с таким номером не найдено!')
           
            
```

```python
def full_list():
    """Возвращает список всех документов"""
    for document in documents:
    
        print(f'{document["type"]} "{document["number"]}""{document["name"]}"')
```

```python
def add_shelf():
    """Добавляет полку с задданым по вводу номером"""
    shelf_num = input('Введите номер полки которую хотите добавить')
    for shelf in directories.keys():
        if shelf_num in shelf:
            print('Такая полка уже существует')
            break
        directories[shelf_num] = []
        print('Полка с номером',shelf_num,'добавлена')
        break
```

```python
def delete_doc():
    """Удаляет документ по заданному номеру"""
    i = 0
    doc_num = input('Введите номер документа')
    for index, document in enumerate(documents):
        if document['number'] == doc_num:
            i = 1
            del(documents[index])
    for document_1 in directories.values():
        if doc_num in document_1:
            document_1.remove(doc_num)
            print('Документ с номером',doc_num,'удален')
            break
    if i == 0:
        print('Документа с таким номером не найдено!')
```

```python
def move_doc():
    """Перемещает указанный документ на заданную полку"""
    i = 0
    s = 0
    doc_num = input('Введите номер документа')
    shelf_num = input('Введите номер полки куда хотите переместить документ')
    for shelf in directories.keys():
        if shelf == shelf_num:
            s = 1
    if s == 1:        
        for document in directories.values():
            if doc_num in document:
                document.remove(doc_num)
                directories[shelf_num].append(doc_num)
                print('Документ с номером',doc_num,'перемещен на полку',shelf_num)
                i = 1
                s = 1
                break
        if i == 0:
            print('Документа с таким номером не существует')
            
    else:
        print('Полки с таким номером не существует')
    
    
```

```python
def add_document():
    """Добавляет новый документ"""
    user_name = input('Введите имя пользователя')
    doc_num = input('Введите номер документа')
    type_doc = input('Введите тип документа')
    shelf_num = input('Введите номер полки на которую хотите добавить ваш документ')
    s = 0
    for shelf in directories.keys():
        if shelf == shelf_num:
            s = 1
    if s == 1:    
        documents.append({'type':type_doc,'number':doc_num,'name':user_name})
        directories[shelf_num].append(doc_num)
        print('Документ добавлен')
    else:
        print('Полки с таким номером не существует')
    
    
```

```python
def main_secretary_function():
    """
    Программа вниательный секретарь. Доступны следующие команды:
    p - people - команда, которая спросит номер документа и выведет имя человека, которому он принадлежит
    s - shelf - команда, которая спросит номер документа и выведет номер полки, на которой он находится
    l - list -команда, которая выведет список всех документов
    as - add shelf - команда, которая спросит номер новой полки и добавит ее в перечень.
    d - delete - команда, которая спросит номер документа и удалит его из каталога и из перечня полок.
    m - move - команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую.
    a - add -команда, которая добавит новый документ в каталог и в перечень полок, спросив его номер, 
    тип, имя владельца и номер полки, на котором он будет храниться.
    q - quit - выход из программы
    """
    while True:
        user_input = input('Добро пожаловать в программу "Внимательный секретарь". Пожалуйста, введите команду')
        if user_input == 'p':
            check_num_doc()
        elif user_input == 's':
            check_shelf()
        elif user_input == 'l':
            full_list()
        elif user_input == 'as':
            add_shelf()
        elif user_input == 'd':
            delete_doc()
        elif user_input == 'm':
            move_doc()
        elif user_input == 'a':
            add_document()
        elif user_input == 'q':
            print('Выходим из программы! До свидания')
            break
        else:
            print('Ошибка! Такой команды не существует')
            break
            
            
#Также вариант добавить break после каждого elif, тогда функция будет сразу выходить из цикла
            
            
            
```
