```python
from pylab import rcParams
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import sklearn
import pylab as pl
```

```python
avocado = pd.read_csv('avocado.csv')
avocado.head()
```

```python
#График которые показывает

pl.figure(figsize=(12,5))
pl.title("Распределение средней цены закупки")
ax = sns.distplot(avocado["AveragePrice"], color = 'b')
```

```python
#также можно сравнить среднее распределение цены авокадо в зависимости от типа
sns.boxplot(y="type", x="AveragePrice", data=avocado, palette = 'pink')
```

```python
#или так
a=sns.catplot(x='type',y='AveragePrice',data=avocado,palette='mako',height=10,kind='boxen')
plt.title('Распределение средней цены по типам Авокадо',size=25)
```

```python
#смотрим зависимость средней цены от объема закупок

plt.figure(figsize=(11,9))
a=sns.regplot(x='Total Volume',y='AveragePrice',data=avocado[avocado['Total Volume']<1500000],color='c')
plt.title('Зависимость средней цены от объема закупок',size=25)
```

```python
#показываем рапределение объема закупок по региону

plt.figure(figsize=(10,15))
a=sns.barplot(x='Total Volume',y='region',data=avocado[avocado.type=='organic'].query('region != "TotalUS"'),palette='coolwarm')
a.set_yticklabels(a.get_yticklabels(),size=16)
plt.title('Итоговый объем закупок по каждому региону',size=30)
```

```python
#Сравниваем среднюю цену по регионам
plt.figure(figsize=(13,15))
a=sns.barplot(x='AveragePrice',y='region',data=avocado,palette='nipy_spectral',hue='type')
a.set_yticklabels(a.get_yticklabels(),size=16)
plt.title('График сравнения средний цены по регионам и типов авокадо',size=30)
```
