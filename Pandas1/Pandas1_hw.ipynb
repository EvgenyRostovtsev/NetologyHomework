```python
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')
joined = ratings.merge(movies, on='movieId', how='left')
ratings_5 = joined[ joined.rating == 5.0 ]
ratings_5['title'].value_counts().head(1)
```

# Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
data = pd.read_csv('power.csv')
data.head()
```

```python
countries = ['Latvia', 'Lithuania', 'Estonia']
categories = [4, 12, 21]
years = [2005, 2006, 2007, 2008, 2009, 2010]
```

```python
filtered_data = data[ data.country.isin(countries) & data.year.isin(years) & data.category.isin(categories) & (data['quantity'] > 0) ]
```

```python
filtered_data.sort_values('quantity', ascending=False)
```

```python
print('Суммарное потребление стран Прибалтики по категориям 4, 12, 21 и за 2005-2010 года равно',filtered_data.quantity.sum())
```
